## Ore Processing Stages ##
<a id='ore_sum'></a>
The process is best explained by the following image provided by Zyfra: 



There are two main stages the ore will go through: flotation and purification. Understanding these processes is critical to understanding how the features relate to the recovery rates.  

_**Flotation:**_

Gold ore mixture (_feed_) is fed into the float banks to obtain a rougher gold concentrate (**first target**) and rougher tails (product residue that contains a low concentration of valuable metals). The tails are visualized on the lower-right side of the chart, as each step in the stages produces this residue. The stability of this process is affected by the _volatile_ and _non-optimal_ physicochemical state of the flotation pulp (a mixture of solid particles and liquid).

`Flotation` has become the most widely used method for extracting minerals from their ores. Specifically, it separates and concentrates ores by altering their surfaces such that they are either repelled or attracted by water. The ores are often coated with a water repellent (chemicals and/or oils) to make them float (gold can often be coated with iron or lead, unknown if that is true in our data, but this could reduce flotability). After the mixture binds to the mineral, finely ground particles remain unwetted and adhere to air bubbles which then float to the surface of the pulp. This forms a froth, which can then be removed. _**Purification:**_

Zyfra's purification stage is not exactly explained, but the process is usually similar to what occurs during flotation, but is usually more refined. Zyfra has informed that the rougher concentrate undergoes two stages of purification, also known as cleaning, which is often the case when refining the rough concentrate. The final concentration is obtained after the stages of cleaning are completed - usually until the desired grade of ore is obtained.

```python
def missing_values_view(df, show_outputs = True):
    missing_values = df.isna().sum().nlargest(15)

    missing_values = pd.DataFrame(
        missing_values.reset_index()
                      .rename(columns={'index': 'column',
                                       0: 'miss_count'})
    )

    missing_values['miss_perc'] = (
        (missing_values['miss_count'] / df.shape[0]) * 100
    ).round(2)

    total_missing_perc = (df.isna().any(axis = 1).sum() / len(df)) * 100

    # remove outputs if set to false
    if show_outputs:
        display(missing_values)
        print(f"Total percentage of rows with missing values: {total_missing_perc.round(2)}%" )
        print('\033[1m' + '<<>>' * 16 + '\n')
    else:
        missing_values = missing_values[~missing_values['column'].str.contains('output|calculation')]
        display(missing_values)
        print(f"Total percentage of rows with missing values: {total_missing_perc.round(2)}%" )
        print('\033[1m' + '<<>>' * 16 + '\n')

    return None
```

We still have a large percentage of rows with missing values in the training dataset. Given that our data is time-based and we know that parameters that are next to each other in time are often similar, we can use time-based methods to fill these missing values. Filling missing values is known as `imputation` and there are many ways to impute missing data.  

For Hourly Time Series Data, a few appropriate methods include: 
1. A simple `Forward / Backward Fill`: this method propagates the last observed value forward (or backward) to fill missing entires. 
2. `K-Nearest Neighbors (KNN)`: This method uses the k-nearest neighboring observations to estimate missing values.
4.  `Linear Interpolation`: This method estimates missing values by interpolating between adjacent data points.

While `KNN imputation` is an intriguing option because it accounts for patterns in similar observations, it is more suitable for datasets with multiple time series or multivariate data. Since our data consists of a single time series, `linear interpolation` is the most appropriate choice. This method will leverage the temporal continuity in our data to estimate missing values effectively. In addition, the changes in our hourly data are gradual rather than abrupt non-linear changes. 

For further support on my choice, I would like to refer to [this](https://www.analyticsvidhya.com/blog/2021/06/power-of-interpolation-in-python-to-fill-missing-values/#:~:text=This%20beginner%E2%80%99s%20tutorial%20is%20about) article. 

```python
# Perform linear interpolation on train and test
# This will hit columns in train that we're not using for ML development, but that is okay. 
for set in sets:
    if set != 'full':
        data[set].interpolate(axis = 0,
                              inplace = True)
```

```python
def plot_stage_hists(metal):
    '''
    This function plots histograms to show the distributions of 
    the concentration of a metal at its four different stages:
    feed, rough concentrate, primary cleaner concentrate, and
    final concentrate.
    
    The histograms are purposefully plotted in a step fashion to 
    see the rise or fall in concentration across each chronological
    stage.
    
    The input is a string in periodic table format. 
    '''
    
    metal = metal.lower()
    
    # Check input is only two characters
    if len(metal) != 2:
        raise ValueError("Metal is in wrong format, please try again.")
        
    
    # Create histograms at each stage
    sns.set_style('darkgrid')
    fig, axs = plt.subplots(nrows = 1, ncols = 4, figsize = (13.5, 5), sharey = True)
    axs = axs.flatten()
    
    plt.suptitle(f"Distribution of {metal.title()} Concentration at each Purification Stage", 
                 fontsize = 15, fontweight = 'bold')
    
    #Initialize variables for loop
    #metals is located below in the section
    stage_cols = [stage[:-2] for stage in metals['au']]
    stage_colors = ['green', 'blue', 'orange', 'red']
    
    for i, stage in enumerate(stage_cols):
        sns.histplot(data = data['train'],
                     y = stage + metal,
                     color = stage_colors[i],
                     ax = axs[i])

        # Create stat labels
        mean_val = data['train'][stage + metal].mean()
        zero_count = (data['train'][stage + metal] == 0).sum()

        # Position stat labels
        y_max = axs[i].get_ylim()[1]
        x_max = axs[i].get_xlim()[1]
        axs[i].text(x_max * 0.8, y_max * 0.9, 
                    f'Mean: {mean_val:.2f}%\nZeros: {zero_count}', 
                    fontweight='bold', ha='center')
        
        axs[i].set_title(stages[i], fontsize = 12, fontweight = 'bold')
        axs[i].set_ylabel(f"Concentration of {metal.title()}(%)", fontsize = 12, fontweight = 'bold')
        axs[i].set_xlabel('')
    
    # Add a centered x-label for the entire figure
    fig.text(0.5, 0.04, 'Count', ha='center', fontsize=12, fontweight='bold')
    
    plt.tight_layout(rect=[0, 0.05, 1, 0.95]) 
    plt.show()
```

```python
def cohens_d(x, y):
    '''
    This function will calculate Cohen's d given two pandas.series 
    of data (unequal length) where quantitative evaluation of their 
    differences is needed.
    
    Please see reference-links #5 for reference material. 
    '''
    # Define variables for formula
    nx = len(x)
    ny = len(y)
    mean_x = x.mean()
    mean_y = y.mean()

    s_pool = np.sqrt((((nx - 1) * ((np.std(x)) ** 2)) + ((ny - 1) * ((np.std(y)) ** 2))) / (nx + ny - 2))
    d = (mean_x - mean_y) / s_pool

    return d
```

```python
def plot_total_dists(dataframe):
    '''
    This function plots the distributions of the total concentrations
    across each purification stage.
    '''

    # Set up figure
    sns.set_style('darkgrid')

    fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize = (9,6))
    axs = axs.flatten()

    plt.suptitle("Distributions of Total Metal Concentrations at each Purification Stage",
                 fontsize = 15, fontweight = 'bold')

    stage_colors = ['green', 'blue', 'orange', 'red']

    for i, col in enumerate(total_columns): #total_columns defined in section below
        sns.histplot(data = dataframe[col],
                     stat = 'count',
                     color = stage_colors[i],
                     ax = axs[i])

        zero_count = (dataframe[col] == 0).sum()

        # Position stat label
        y_max = axs[i].get_ylim()[1]
        x_max = axs[i].get_xlim()[1]
        axs[i].text(x_max * 0.2, y_max * 0.9, 
                    f'Zeros: {zero_count}', 
                    fontweight='bold', ha='center')
        
        axs[i].set_title(stages[i], fontsize = 13, fontweight = 'bold')
        axs[i].set_xlabel("Concentration of Metals (%)", fontsize = 10, fontweight = 'bold')
        axs[i].set_ylabel("Count", fontsize = 10, fontweight = 'bold')

    plt.tight_layout()
    plt.show()
```

```python
def check_threshold(threshold, stats = False):
    '''
    This function will checck the final recovery values
    when we set our threshold to a certain value, so that 
    we may eliminate unwanted anomalies.
    '''
    
    for col in total_columns:
        check_condition = data['train'][col] < threshold
        min_val = data['train'][~check_condition][col].min()
        recov_at_min = data['train']['final.output.recovery'][data['train'][col] == min_val].values[0]
        n_anoms = data['train'][check_condition][col].shape[0]
        perc_anoms = n_anoms / len(data['train'][col]) * 100

        if stats:
            print(col)
            print(f"Number of anomalies: {n_anoms}")
            print(f"Percentage of anomalies: {perc_anoms:.2f}%\n")
        else:
            print(f"{col}'s minimum value after threshold: {min_val:.2f}")
            print(f"Final recover at this value: {recov_at_min:.2f}\n")
```

## 4.2 Distributions of 'Feed Size' in Training Set Vs. Test Set ##
<a id='feed_comp'></a>

To perform this analysis we will complete three steps: 
1. Extract the `feed_size` columns from their respective datasets.
2. Visualize the distributions using a Box plot.
3. Perform a statistical test to quantitatively compare the distributions.

```python
# Step 1 - Extract data
train_feed = data['train']['rougher.input.feed_size']
test_feed = data['test']['rougher.input.feed_size']

# create combined dataframe
combined_feed = pd.DataFrame({
    'Feed Particle Size': pd.concat([train_feed, test_feed]),
    'Dataset': ['Train'] * len(train_feed) + ['Test'] * len(test_feed) 
})

combined_feed.sample(5)

# Step 2 - build boxplot
sns.set_style('darkgrid')
plt.figure(figsize = (10,6))
sns.boxplot(data = combined_feed, 
            x = 'Dataset', y = 'Feed Particle Size',
            hue = 'Dataset')

# Calculate mean
mean_vals = (
    (combined_feed.query("Dataset == 'Train'")['Feed Particle Size']).mean().round(2), 
    (combined_feed.query("Dataset == 'Test'")['Feed Particle Size']).mean().round(2)
)

median_vals = (
    round((combined_feed.query("Dataset == 'Train'")['Feed Particle Size']).median(), 2),
    round((combined_feed.query("Dataset == 'Test'")['Feed Particle Size']).median(), 2)
)

# apply labels
plt.text(-0.3, 400,
         f"Train Mean:\n{mean_vals[0]}\n\nTrain Median:\n{median_vals[0]}", 
         ha = 'center', fontweight = 'bold')
plt.text(.7, 400,
         f"Test Mean:\n{mean_vals[1]}\n\nTest Median:\n{median_vals[1]}", 
         ha = 'center', fontweight = 'bold')

plt.title("Feed Particle Size in Train Vs. Test Sets - Box Distribution",
          fontsize = 15, fontweight = 'bold')
plt.xlabel('Dataset', 
           fontsize = 12, fontweight = 'bold')
plt.ylabel('Feed Particle Size', 
           fontsize = 12, fontweight = 'bold')
plt.show()
```

The distributions are very close with the median and means being apart by about an interval of 5 - the test mean and median being slightly lower than the train. We will perform a `Kolmogorov-Smirnov Test` to quantitatively assess if the distributions are different, in addition to the visual above. 

The KS Test is a non-parametric, distribution-free test; making the test popular for checking normality, but it can also be used to compare two distributions to see if they are the same. Luckily for us, it is very easy to apply!

```python
# Step 3 - Perform KS test
ks_stat, p_value = ks_2samp(train_feed, test_feed)

print('\033[1m')
print(f"D Statistic: {ks_stat:.4f}, p-value: {p_value}")
print('\033[0m')

if p_value < 0.05:
    print("The distributions of feed particle sizes are significantly different.")
else:
    print("The distributions of feed particle sizes are not significantly different.")
```

The KS Test has detected a significant difference, which might actually be due to its high sensitivity to even the slightest of differences when the sample sizes are large. This highlights the potential for _statistical significance_ without _practical significance_. The other problem with relying on some statistical tests for the comparison of these distributions, are that we are not looking to compare a specific value to define the distribution (such as the mean or median), but are looking to assess the shapes of the whole distribution. Thus, we need to better visualize the difference and so we will turn to the ECDF plots.

The `ECDF plot`, per Seaborn documentation, "represents the proportion or count of observations falling below each unique value in a dataset." This allows us to ensure each observation is visualized directly. The downside to this type of chart is that it doesn't tell us much of anything else, only how similar the shapes of the distributions are and where.  

```python
# Plot ECDFs
sns.set_style('darkgrid')
plt.figure(figsize=(12, 6))

sns.ecdfplot(data = train_feed, label = 'Train ECDF', linestyle = '--')
sns.ecdfplot(data = test_feed, label = 'Test ECDF', linestyle = '--')

plt.title('Empirical Cumulative Distribution Functions (ECDF)', 
          fontsize=15, fontweight = 'bold')
plt.xlabel('Feed Particle Size', fontsize = 12, fontweight = 'bold')
plt.ylabel('ECDF', fontsize = 12, fontweight = 'bold')
plt.ylim(-0.01, 1.01)
plt.legend()
plt.grid(True)
plt.show()
```

The plots are very similar which indicates that the distribution of feed particle sizes between the two sets are similar. The differences between the two distributions is highly likely negligable. While the effect is likely small, we can calculate it using `Cohen's d`, this way we can quantitaively define the difference afterall. 

```python
# Use function to calculate cohen's d
d = cohens_d(train_feed, test_feed)
print(f"Cohen's d: {d:.4f}")

if np.abs(d) < 0.2:
    print("Cohen's d indicates a small effect size, suggesting no meaningful difference.")
else:
    print("Cohen's d indicates a medium or large effect size, suggesting a meaningful difference.")
```

**In conclusion, the analysis suggests that the distributions between the `test particle feed size` and the `train particle feed size` are similar enough andd should not harm the models' development.**

The functions below, some based on previous projects, will help automate tasks in this section. 

```python
def sMAPE(y_true, y_pred):
    '''
    Function to calculate the sMAPE of a model.
    
    **Unsure as to why  it's returning negative, but added a - sign 
    to correct this.**
    '''

    epsilon = 1e-10
    n = len(y_true)
    nume = np.abs(y_true - y_pred)
    denom = (np.abs(y_true) + np.abs(y_pred)) / 2
    sMAPE = (100/n) * sum(nume / (denom + epsilon))

    return sMAPE # kept returning negative for some reason, formula should be correct
```

```python
def final_sMAPE(y_true, y_pred):
    '''
    This function calculates the final sMAPE of a model.
    '''
    
    # Split true values into rougher and final recovery
    y_true_r = y_true[:,0] # all rows of first column of array
    y_true_f = y_true[:,1] # all rows in second column of array

    # Split predicted values in same fashion
    y_pred_r = y_pred[:,0]
    y_pred_f = y_pred[:,1]

    # Calculate sMAPE for both 
    sMAPE_r = sMAPE(y_true_r, y_pred_r)
    sMAPE_f = sMAPE(y_true_f, y_pred_f)

    # Perform calculation
    final_sMAPE = 0.25 * sMAPE_r + 0.75 * sMAPE_f

    return -final_sMAPE
```

```python
def best_model_search(feat_train, targ_train, estimator, param_grid, multi_output = True):
    '''
    This function is based off the function from project 8, so please see project 8 if needed. This function
    will aid in determining the hyperparameters/best model for the estimator chosen in the parameters. This function
    will also return a dataframe containing details of the results from each model as well as the 
    best model with the lowest final sMAPE percentage. 

    Parameters:
        feat_train (df)          - dataframe containing the features data
        targ_train (series)      - series containing the target data
        estimator (model object) - model object chosen to train and predict
        param_grid (dict)        - dictionary containing all of the parameters for your chosen estimator
        multi_output (bool)      - boolean, applies MultiOutputRegressor if estimator requires it
    '''

    # Declare the variables needed to store information
    model_param_list = []
    sMAPE_list = []
    best_model = {}
    best_model_data = {'final_sMAPE': float('inf'),
                       'hyperparameters': None}

    # Perform search for best model (the loop)
    keys, values = zip(*param_grid.items())

    for param_value in product(*values):
        model_params = dict(zip(keys, param_value))

        # Apply multi-output and declare model
        if multi_output:
            model = MultiOutputRegressor(estimator(**model_params))
        else:
            model = estimator(**model_params)
        
        # Train the model and find metrics            
        model.fit(feat_train, targ_train)

        # Score the model
        scores = cross_val_score(model, feat_train, targ_train, cv = 10,
                                 scoring = f_sMAPE_score, n_jobs = -1)
        
        score = scores.mean()
        
        # Store the information
        model_param_list.append(model_params)
        sMAPE_list.append(score)
        
        # Check if best model
        if score < best_model_data['final_sMAPE']:
            best_model = model
            best_model_data['hyperparameters'] = model_params
            best_model_data['final_sMAPE'] = score

    # Create DataFrame to store each result
    results_df = pd.DataFrame({
        'params': model_param_list,
        'final_sMAPE': sMAPE_list
    })
    # Unpack the dict in the columns
    results_df = pd.concat(
        [
            pd.json_normalize(results_df['params']),
            results_df[['final_sMAPE']]
        ], 
        axis=1)
    results_df = results_df.sort_values(by = 'final_sMAPE').reset_index(drop=True)

    print("Completed!")

    return results_df, best_model, best_model_data
```

## 6.1 Training Multiple Models ##
<a id='training'></a>

Please see the [Formulas](#forms) section for the formula that will be used to evaluate the models. We will be using `sMAPE`, or symmetric Mean Absolute Percentage Error. It is similar to MAE, but is expressed in relative values instead of absolute ones. It equally takes into account the scale of both the target and prediction. In order to truly measure the models performance for this project, we will need to calculate sMAPE for both the rougher recovery and the final recovery, and then use the `final sMAPE` formula. 

In order to use time most efficiently, we will first test multiple models without any hyperparameter tuning and see how they perform via `cross validation` and their `final sMAPE` measurements. From this list, we will fine-tune the top 2-3 models to achieve the smallest `final sMAPE` possible. Lastly, we will test our final model on the test dataset. 

```python
# Dictionary to hold all models
models = {}

# All models to be tested; excited to try some new ones
models['Baseline'] = DummyRegressor(strategy = 'mean')
models['Linear Regression'] = MultiOutputRegressor(LinearRegression())
models['KNN'] = MultiOutputRegressor(KNeighborsRegressor())
models['Decision Tree'] = DecisionTreeRegressor(max_depth = 10, random_state = 12345)
models['Random Forest'] = RandomForestRegressor(max_depth = 10, random_state = 12345, n_jobs = -1)
models['Gradient Boosting'] = MultiOutputRegressor(GradientBoostingRegressor(random_state = 12345))
models['Linear SVR'] = MultiOutputRegressor(SVR(kernel = 'linear'))
models['Non-Linear SVR'] = MultiOutputRegressor(SVR(kernel = 'rbf'))
models['XGB'] = MultiOutputRegressor(XGBRegressor(random_state = 12345))
# prevent the logs with verbose = 0 (CatBoost)
models['CatBoost'] = MultiOutputRegressor(CatBoostRegressor(verbose = 0, 
                                                            random_seed = 12345))
```

```python
# Establish evalutation metric via custom scorer
f_sMAPE_score = make_scorer(final_sMAPE, greater_is_better = False)
```

```python
model_names = []
model_sMAPE = []

for model_name, model in models.items():

    # Announce which model is in training
    start_time = time.time()
    print(f"Currently training {model_name}")
    
    # Score the model via CV
    scores = cross_val_score(model, X_train, Y_train, cv = 10,
                           scoring = f_sMAPE_score, n_jobs = -1)

    score = scores.mean()

    #Append the data to lists
    model_names.append(model_name)
    model_sMAPE.append(score)

    # Announce success
    end_time = time.time()
    print(f"Training successful. Time taken: {end_time - start_time:.0f} seconds.\n")


# Exit Loop
# Data frame to store information and sort it.
result_df = pd.DataFrame({
    'model': model_names,
    'final_sMAPE': model_sMAPE
}).sort_values(by = 'final_sMAPE').reset_index(drop = True)
```

## 6.3 Evaluation of the Final Model on Test Set ##

```python
# Assign the final model
final_model = models['Tuned CatBoost']

# train
final_model.fit(X_train, Y_train)

# Evaluate the model
Y_pred_final = final_model.predict(X_test)
final_score = final_sMAPE(Y_test, Y_pred_final)
final_score = -final_score

# Observe the result
print(f"The final sMAPE score of the fine-tuned CatBoost model on the test set is: {final_score:.2f}%")
```

```python
# Check baseline
base_sanity = models['Baseline']

# train
base_sanity.fit(X_train, Y_train)

# Evaluate
bs_pred = base_sanity.predict(X_test)
bs_score = -final_sMAPE(Y_test, bs_pred)

# Observe
print(f"The final sMAPE score of the Baseline model on the test set is: {bs_score:.2f}%.")
```

```python
# Print a comparison (curious)
ratio = (bs_score - final_score) / bs_score
ratio *= 100
print(f"The fine-tuned Catboost model outperformed the baseline model by {ratio:.2f}%.")
```

**The fine-tuned CatBoost model obtained a final sMAPE score of 8.49% on the test set and outperformed the baseline model by 13.7%.**

# 7.0 Conclusion #
<a id='conclusion'></a>
_________________________________________________________________________________________________________________________________________________________

**Data Exploration and Preprocessing:**

In this stage, the following issues were identified and resolved (_not all resolutions occurred in this stage_):
- The `date` column was identified to be the wrong data-type, and was converted to `datetime`.
- A large analysis was conducted on the missing values found throughout the datasets. To resolve this:
    - dropped missing values in target columns in `train dataset`
    - added/merged target data from `full set` to `test set` and dropped missing values in target columns in `test set`.
    - filled all remaining missing values using `interpolation`
- Features identified to very significantly in scale and so a StandardScaler was applied.

**Exploratory Data Analysis:**

- Observed the distributions of the concentrations of each metal across each purification stage via a bar chart, and horizontal histograms:
    - `Gold` increased
    - `Silver` increased then decreased
    - `Lead` increased
- Through these distributions we observed a large amount of 0 values, or values close to zero
- Compared the distributions of feed size amongst the training and test sets via a box plot. While close, there was no quantitave determination of difference. A `KS-test` was performed and concluded the distributions are significantly different; however, this could be due to the tests limitations so the ECDF plots were compared. These plots were determined to be highly similar and quantified numericaly via the affect size using `Cohen's d` which returned a difference percentage < 20%: concluding to be a minimal difference. Therefore, we concluded the distributions to be similar enough to not harm the ML algorithms.
- Used distributions of the total concentrations of each metal across the purification stages to investigate the possible zero-value anomalies discovered earlier. We found impossible instances at these values where the recovery is greater than the concentrations. We found the optimal threshold for the anomalies and removed them from the training set.

**Development of Models**

- A total of 10 models were tested with no hyperparamters. The top 3 models were then tuned via their hyperparameters. 2/3 were tested extensibly, the gradient boosting was not tested as extensively due to the run-time.
- The top 3 performing models were: `Non-linear SVR`, `CatBoost`, and `Gradient Boosting`.
- The `CatBoost` model was chosen as the final model after having the best final sMAPE scores after tuning hyperparameters.
- The `Fine-tuned CatBoost` model obtained a final sMAPE score of **8.49%** on the test setand outperformed the baseline by 13.7%.
